## Task 5 – Introduction to Sets

Create a list containing the following IP addresses:

192.168.1.1  
192.168.1.2  
192.168.1.1  
10.0.0.5  
10.0.0.5  

Convert this list into a **set**.

Print the set.

```python
ips = ["192.168.1.1", "192.168.1.2", "192.168.1.1", "10.0.0.5", "10.0.0.5"]
new_ips_in_set = set(ips)
print(new_ips_in_set)
```

## Task 6 – Unique Values

Using the set created in the previous task:

- Print each IP address on a separate line
- Print how many unique IP addresses exist

```python
for item in new_ips_in_set:
    print(f"{item}")
print(f"total unique ips: {len(new_ips_in_set)}")
```

## Task 7 – User Input with Lists

Ask the user to enter **five IP addresses**.

Store them in a list.

Then:

- Convert the list into a set
- Print the unique IP addresses
- Print the total number of unique IP addresses

```python
ips = []

for i in range(5):
    ip = input(f"Please enter {i + 1}. ip address: ")
    ips.append(ip)


print(f"Entered ips = {ips}")
ips_in_set = set(ips)
print(f"Unique ips = {ips_in_set}")
print(f"Total number of unique ips = {len(ips_in_set)}")
```
